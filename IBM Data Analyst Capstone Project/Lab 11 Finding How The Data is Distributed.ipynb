### Step 3: Handle Missing Data

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

```python
## Write your code here
exp_dict = {'Less than 1 year' : '1', 'More than 50 years' : '51'}
           
df['ExperienceLevel'] = df['YearsCodePro'].replace(exp_dict)
df['ExperienceLevel'] = df['ExperienceLevel'].astype(int)

df.ExperienceLevel.value_counts() 
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
sizes = df.JobSat.value_counts()
labels = list(df.JobSat.unique())
colors = ['yellowgreen','red','gold','lightskyblue','white','lightcoral','blue','pink', 'black','yellow', 'cyan']
porcent = 100.*sizes/sizes.sum()
patches, texts = plt.pie(sizes, colors=colors, startangle=90, radius=1.2)
labels = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(labels, porcent)]

sort_legend = True
if sort_legend:
    patches, labels, dummy =  zip(*sorted(zip(patches, labels, sizes),
                                          key=lambda x: x[2],
                                          reverse=True))

plt.legend(patches, labels, loc='center right', bbox_to_anchor=(-0.1, 1.),
           fontsize=8)
# Add a title
plt.title("Job Satisfaction Distribution")

# Display the chart
plt.show()
```

```python
sns.kdeplot(data=df, x="JobSat", fill=True)
plt.title("KDE Plot of Job Satisfaction")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sizes.sort_index().plot(kind='bar')
plt.title("Distribution of Job Satisfaction")
plt.xlabel("Job Satisfaction")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
!pip install matplotlib-venn
```

```python
from matplotlib_venn import venn2
# Convert columns to sets
set_a = set(df['LanguageHaveWorkedWith'])
set_b = set(df['LanguageWantToWorkWith'])

# Calculate the sizes for the Venn diagram
# Ab: elements in A, not in B
Ab = len(set_a - set_b)
# aB: elements in B, not in A
aB = len(set_b - set_a)
# AB: elements common to A and B
AB = len(set_a.intersection(set_b))

# Create the Venn diagram
venn2(subsets=(Ab, aB, AB), set_labels=('LanguageHaveWorkedWith', 'LanguageWantToWorkWith'))
plt.title("Venn Diagram of Language to Have Worked With and Language Want to Work With")
plt.show()
```

```python
colA = []
for val in df.LanguageHaveWorkedWith: 
    s = val
    parts = s.split(';')
    colA.append(parts)

flattened_colA = [item for sublist in colA for item in sublist]
```

```python
colB = []
for val in df.LanguageWantToWorkWith: 
    s = val
    parts = s.split(';')
    colB.append(parts)

flattened_colB = [item for sublist in colB for item in sublist]
```

```python
from matplotlib_venn import venn2
# Convert columns to sets
set_a = set(flattened_colA)
set_b = set(flattened_colB)

# Calculate the sizes for the Venn diagram
# Ab: elements in A, not in B
Ab = len(set_a - set_b)
# aB: elements in B, not in A
aB = len(set_b - set_a)
# AB: elements common to A and B
AB = len(set_a.intersection(set_b))

# Create the Venn diagram
venn2(subsets=(Ab, aB, AB), set_labels=('LanguageHaveWorkedWith', 'LanguageWantToWorkWith'))
plt.title("Venn Diagram of Language to Have Worked With and Language Want to Work With")
plt.show()
```

```python
from collections import Counter
# Count frequencies
count_work_with = Counter(flattened_colA)
count_want_with = Counter(flattened_colB)

# Combine into DataFrame
languages = sorted(set(flattened_colA + flattened_colB))
df2 = pd.DataFrame({
    'Language': languages,
    'Currently Work With': [count_work_with.get(lang, 0) for lang in languages],
    'Want to Work With': [count_want_with.get(lang, 0) for lang in languages]
})

# Plot grouped bar chart
plt.figure(figsize=(12, 6))
x = range(len(df2))
bar_width = 0.35

plt.bar([i - bar_width/2 for i in x], df2['Currently Work With'], width=bar_width, label='Currently Work With')
plt.bar([i + bar_width/2 for i in x], df2['Want to Work With'], width=bar_width, label='Want to Work With')

plt.xticks(x, df2['Language'], rotation=45, ha='right')
plt.ylabel('Count')
plt.title('Comparison of Programming Languages: Work With vs Want to Work With')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Create a dictionary mapping each country to its continent
continent_map = {}

for c in nalist:
    continent_map[c] = 'North America'
for c in salist:
    continent_map[c] = 'South America'
for c in eulist:
    continent_map[c] = 'Europe'
for c in asialist:
    continent_map[c] = 'Asia'
for c in africalist:
    continent_map[c] = 'Africa'
for c in oceanialist:
    continent_map[c] = 'Oceania'
for c in islandlist:
    continent_map[c] = 'Oceania'  

# Map the continent to the dataframe
df['Continent'] = df['Country'].map(continent_map)
df
```

```python
# Count distribution
mode_counts = df.groupby(['Continent', 'RemoteWork']).size().unstack(fill_value=0)

# Reuse the grouped data
plt.figure(figsize=(8, 5))
sns.heatmap(mode_counts, annot=True, cmap='Blues', fmt='d')
plt.title('Heatmap of Work Mode Distribution by Region')
plt.xlabel('Work Mode')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

```python
# Plot grouped bar chart
mode_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Distribution of Work Mode by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Work Mode')
plt.tight_layout()
plt.show()
```
