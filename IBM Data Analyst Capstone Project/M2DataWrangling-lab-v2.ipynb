### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
df.Country = df.Country.str.lower() 

country_dict = {'united states of america': 'usa', 
                'united kingdom of great britain and northern ireland' : 'uk', 
                'iran, islamic republic of...' : 'iran',
                'republic of north macedonia' : 'north macedonia',
                'russian federation' : 'russia', 
                'united republic of tanzania' : 'tanzania', 
                'venezuela, bolivarian republic of...' : 'venezuela',
                'republic of moldova' : 'moldova', 
                'united arab emirates' : 'uae', 
                'hong kong (s.a.r.)' : 'hong kong',   
                "lao people's democratic republic" : 'laos', 
                'democratic republic of the congo' : 'dr congo',
                'syrian arab republic': 'syria', 
                "côte d'ivoire" : 'cdi', 
                "democratic people's republic of korea" : 'north korea',
                'congo, republic of the...' : 'congo',  
                'libyan arab jamahiriya' : 'libya', 
                'brunei darussalam' : 'brunei',  
                'micronesia, federated states of...' : 'micronesia'}  
                

df.loc[df.Country == 'viet nam', 'Country'] = 'vietnam'
df['Country'] = df['Country'].replace(country_dict)
df.Country.unique()
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
# Country done above 
edlevel_dict = {'Primary/elementary school' : 'Primary', 
                'Bachelor’s degree (B.A., B.S., B.Eng., etc.)' : 'Bachelor',
               'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)' : 'Masters', 
               'Some college/university study without earning a degree' : 'CourseUnits', 
               'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)' : 'GED', 
               'Professional degree (JD, MD, Ph.D, Ed.D, etc.)' : 'Professional',
                'Associate degree (A.A., A.S., etc.)' : 'Associate', 
                'Something else' : 'Other'}
                
df['EdLevel'] = df['EdLevel'].replace(edlevel_dict)
df['EdLevel'].fillna('NA', inplace=True)
df.EdLevel.unique() 
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df.isnull().sum())
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here 
numerical_cols = df.select_dtypes(include='number').columns.tolist()
numerical_cols.remove('ResponseId') 
numerical_cols 
```

```python
for c in numerical_cols: 
    df[c] = df[c].fillna(df[c].mean())
    print(c, df[c].isna().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
object_columns_df = df.select_dtypes(include=['object'])
cat_cols = object_columns_df.columns
for c in cat_cols: 
    df[c] = df[c].fillna(df[c].mode()[0])
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
# Find the minimum and maximum values in 'ConvertedCompYearly'
min_value = float(df['ConvertedCompYearly'].min())
max_value = float(df['ConvertedCompYearly'].max())

# Apply Min-Max normalization to 'ConvertedCompYearly'
df['ConvertedCompYearly_MinMax'] = (df['ConvertedCompYearly'].astype(float) - min_value) / (max_value - min_value)
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
import numpy as np 
df['Log_ConvertedCompYearly'] = np.log(df['ConvertedCompYearly'] + 1)
df
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
exp_dict = {'Less than 1 year' : '1', 'More than 50 years' : '50'}
           
df['ExperienceLevel'] = df['YearsCodePro'].replace(exp_dict)
df['ExperienceLevel'] = df['ExperienceLevel'].astype(int)
df.ExperienceLevel.unique()
```
